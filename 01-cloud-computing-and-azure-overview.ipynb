**Microsoft Azure** is a cloud computing platform that lets you build, deploy, and manage applications across a global network of data centers — without owning any hardware.

Like all cloud providers, Azure follows a simple principle: instead of buying and maintaining your own servers, you rent compute, storage, networking, and dozens of other services on demand — and pay only for what you use.

## Traditional IT vs. Cloud Computing

| | Traditional IT | Cloud (Azure) |
|---|---|---|
| **Hardware** | Buy and own servers | Rent on-demand |
| **Capacity planning** | Over-provision for peak | Scale up/down as needed |
| **Provisioning speed** | Weeks | Minutes |
| **Cost model** | Large upfront CapEx | Pay-as-you-go OpEx |
| **Maintenance** | Your team | Microsoft manages the infrastructure |
| **Global reach** | Expensive and slow | Deploy to 60+ regions in minutes |

### The old pain points
- Paying for idle capacity — servers sitting at 10–20% utilisation
- Long procurement cycles — ordering hardware takes weeks or months
- Guessing capacity — too little causes outages, too much wastes budget
- Running physical data centers — racks, cooling, power, physical security

## What is Microsoft Azure?

Azure launched in **2010** as "Windows Azure" and was rebranded to **Microsoft Azure** in 2014. It is Microsoft's public cloud platform and the **second-largest cloud provider** in the world (after AWS).

Key facts:
- **200+ services** spanning compute, storage, databases, AI/ML, networking, security, DevOps, and more
- Available in **60+ regions** across the globe — more than any other cloud provider
- Deeply integrated with **Microsoft's ecosystem** — Active Directory, Windows Server, Office 365, GitHub, Visual Studio
- Strong in **enterprise and hybrid** scenarios, with tools like Azure Arc for managing on-premises resources through Azure

Azure is the default cloud choice for organisations already invested in the Microsoft ecosystem.

## Cloud Deployment Models

### ☁️ Public Cloud
Resources are owned and operated by a provider (Azure, AWS, GCP) and shared across customers over the internet.  
You never see or touch the hardware. This is the most common model.

### 🏢 Private Cloud
Cloud infrastructure dedicated to a single organisation — either on-premises or hosted by a provider.  
More control and isolation, but more responsibility and cost. Azure Stack Hub lets you run Azure services in your own data center.

### 🔀 Hybrid Cloud
A mix of public and private cloud, connected together.  
Azure excels here — **Azure Arc** lets you manage on-premises servers, Kubernetes clusters, and databases from the Azure portal as if they were Azure resources.

### 🤝 Multi-Cloud
Using multiple public cloud providers simultaneously.  
Azure Arc also supports managing workloads running on AWS or GCP from a single Azure control plane.

## Cloud Service Models

| Model | Full Name | You manage | Azure Examples |
|---|---|---|---|
| **IaaS** | Infrastructure as a Service | OS, runtime, app, data | Azure VMs, Virtual Network, Managed Disks |
| **PaaS** | Platform as a Service | App and data only | App Service, Azure SQL, Azure Functions |
| **SaaS** | Software as a Service | Nothing — just use it | Microsoft 365, Dynamics 365, Power BI |

A useful way to think about it: as you move from IaaS → PaaS → SaaS, Microsoft takes over more of the stack and you focus more on your application logic.

## Shared Responsibility Model

In the cloud, security and management responsibilities are **shared** between you and Microsoft.

| Responsibility | On-premises | IaaS | PaaS | SaaS |
|---|---|---|---|---|
| Physical data center | You | **Microsoft** | **Microsoft** | **Microsoft** |
| Network & hardware | You | **Microsoft** | **Microsoft** | **Microsoft** |
| Operating system | You | You | **Microsoft** | **Microsoft** |
| Application | You | You | You | **Microsoft** |
| Identity & data | You | You | You | You |

> **You always own your data and identities** — regardless of service model.

## Key Benefits of Azure

1. **High availability and reliability**  
   Azure offers SLAs (service-level agreements) of up to 99.99% uptime, with redundancy built across availability zones.

2. **Scalability and elasticity**  
   Scale resources up/down or out/in automatically in response to demand — no manual intervention needed.

3. **Global reach**  
   60+ regions let you deploy close to your users, meet data residency requirements, and build geo-redundant architectures.

4. **Security and compliance**  
   Microsoft invests over \$1B/year in security. Azure meets 100+ compliance certifications — ISO 27001, SOC 2, GDPR, HIPAA, and more.

5. **Cost management**  
   Pay-as-you-go, reserved instances for predictable workloads, and Azure Cost Management for visibility and budgeting.

6. **Hybrid integration**  
   Best-in-class hybrid capabilities — Azure Arc, Azure Stack, ExpressRoute for private connectivity from on-premises.

## How to Interact with Azure

### Azure Portal
Web-based GUI at [portal.azure.com](https://portal.azure.com). Good for exploration, monitoring, and one-off tasks.

### Azure CLI
Cross-platform command-line tool. Ideal for scripting and automation.
```bash
az login
az group create --name my-rg --location eastus
az vm list --output table
```

### Azure PowerShell
PowerShell module for Azure — preferred in Windows-heavy or enterprise environments.
```powershell
Connect-AzAccount
New-AzResourceGroup -Name my-rg -Location eastus
```

### Azure SDKs
Programmatic access from your application code. Available for Python, JavaScript, Java, .NET, Go, and more.
```python
from azure.identity import DefaultAzureCredential
from azure.mgmt.resource import ResourceManagementClient

credential = DefaultAzureCredential()
client = ResourceManagementClient(credential, subscription_id)
```

### Azure Resource Manager (ARM)
The underlying control plane for all Azure operations. Everything — portal clicks, CLI commands, SDK calls — goes through ARM. It handles authentication, authorisation, and resource lifecycle management.
