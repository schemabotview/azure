Azure operates one of the largest cloud infrastructures in the world — spanning **60+ regions** across every major continent. It is architected for high availability, fault tolerance, and compliance, with multiple layers of physical separation that protect your workloads from hardware failures, natural disasters, and regulatory boundary violations.

## Geographies

At the top level, Azure organises its infrastructure into **Geographies** — large areas that typically correspond to a country or group of countries sharing the same data residency and compliance requirements.

Examples:
- **United States** — contains regions like East US, West US, Central US
- **Europe** — contains North Europe (Ireland), West Europe (Netherlands), UK South, Germany West Central
- **Asia Pacific** — contains East Asia (Hong Kong), Southeast Asia (Singapore), Australia East
- **India** — contains Central India, South India, West India

Each geography contains **at least two regions** so that customers can meet data residency requirements while also enabling disaster recovery.

> Geographies are a compliance and sovereignty boundary — data stays within the geography unless you explicitly move it.

## Regions

A **Region** is a set of data centers deployed within a defined geographic area, connected by a dedicated low-latency network.

Azure has **60+ regions** — more than any other cloud provider — across North America, South America, Europe, Africa, Middle East, Asia Pacific, and Australia.

Example regions:

| Region Name | Location |
|---|---|
| East US | Virginia, USA |
| West Europe | Netherlands |
| UK South | London, UK |
| Southeast Asia | Singapore |
| Australia East | New South Wales, AU |
| Central India | Pune, India |
| Brazil South | São Paulo, Brazil |

### How to choose a Region

| Factor | Consideration |
|---|---|
| **Compliance & data residency** | Regulations may require data to stay within a country or region |
| **Latency** | Deploy close to your end users |
| **Service availability** | Not every Azure service is in every region — check the products table |
| **Pricing** | Costs vary slightly between regions |
| **Region pair** | Choose a region that has a designated pair for disaster recovery |

## Availability Zones

Within a region, Azure divides infrastructure into **Availability Zones (AZs)** — physically separate locations, each with independent power, cooling, and networking.

- Each zone is one or more data centers
- Zones within a region are connected by **high-speed private fibre** (< 2ms round-trip latency)
- Regions that support AZs have a **minimum of 3 zones**
- Zones are numbered: Zone 1, Zone 2, Zone 3

### Zonal vs. Zone-Redundant services

| Type | Behaviour | Example |
|---|---|---|
| **Zonal** | You pin the resource to a specific zone | VMs, Managed Disks |
| **Zone-redundant** | Azure automatically replicates across zones | Azure SQL Database, Storage |
| **Non-regional** | Resilient by design — no zone concept needed | Azure Active Directory, Traffic Manager |

> Not all regions support Availability Zones. Always verify before designing for zone redundancy.

### Why AZs matter

Deploying across multiple Availability Zones protects against a **single data centre failure**. If Zone 1 goes down, your application continues running in Zones 2 and 3. This is the foundation of **high availability** within a region.

## Region Pairs

This is a concept unique to Azure. Every Azure region is **paired with another region** within the same geography, at least 300 miles apart. Together they form a **Region Pair**.

Example pairs:

| Primary Region | Paired Region |
|---|---|
| East US | West US |
| UK South | UK West |
| North Europe (Ireland) | West Europe (Netherlands) |
| Southeast Asia (Singapore) | East Asia (Hong Kong) |
| Brazil South | South Central US |

### Why Region Pairs matter

- **Planned maintenance**: Azure staggers updates across pairs — your paired region is never updated at the same time, so one is always available
- **Outage recovery**: If a region-wide outage occurs, Microsoft prioritises restoring one region from each pair
- **Data replication**: Services like Azure Storage offer geo-redundant replication that automatically replicates to the paired region
- **Compliance**: Paired regions stay within the same geography, so data doesn't cross sovereignty boundaries

> When designing disaster recovery, you should architect your primary and secondary workloads in a **region pair**, not just any two regions.

## Availability Sets

Before Availability Zones existed, Azure offered **Availability Sets** for protecting VMs against hardware failures *within a single data centre*.

An Availability Set groups VMs across two dimensions:

| Concept | What it protects against | How |
|---|---|---|
| **Fault Domain (FD)** | Physical hardware failure (shared power or network switch) | VMs spread across up to 3 fault domains |
| **Update Domain (UD)** | Planned maintenance reboot | VMs spread across up to 20 update domains — only one UD rebooted at a time |

```
Availability Set
├── Fault Domain 0  →  VM-1, VM-4  (separate rack / power)
├── Fault Domain 1  →  VM-2, VM-5
└── Fault Domain 2  →  VM-3, VM-6
```

> **Availability Sets vs. Availability Zones**: Availability Sets protect within a single data centre (no SLA for zone-level failure). Availability Zones protect across physically separate buildings. For new architectures, **prefer Availability Zones** where supported.

## Azure Edge Network

Beyond regions and zones, Azure has a global network of **edge locations** used to accelerate content delivery and reduce latency for end users.

| Component | Purpose |
|---|---|
| **Azure CDN** | Caches static content at edge nodes close to users |
| **Azure Front Door** | Global HTTP load balancer with built-in CDN, WAF, and routing |
| **Azure DNS** | Anycast DNS with nodes distributed globally |
| **Azure DDoS Protection** | Traffic scrubbing at the network edge |

### Azure Edge Zones

**Azure Edge Zones** (and **Azure Edge Zones with carriers**) extend Azure compute into metropolitan areas and telecom networks — similar to AWS Local Zones and Wavelength Zones.

- **Azure Edge Zones**: Bring Azure services to densely populated cities not yet served by a full region
- **Azure Edge Zones with carriers**: Embed Azure compute inside 5G telecom networks for ultra-low latency (single-digit milliseconds) — targeted at gaming, AR/VR, autonomous vehicles, and IoT

## Sovereign & Dedicated Regions

Some governments and regulated industries require cloud infrastructure that is physically isolated from the public Azure cloud.

### Azure Government (US)
- Physically separate datacenters operated by **screened US citizens only**
- Meets FedRAMP High, DoD IL2/IL4/IL5, ITAR, and other US federal compliance standards
- Regions: US Gov Virginia, US Gov Arizona, US Gov Texas

### Azure China (21Vianet)
- Operated by **21Vianet**, a separate Chinese company — Microsoft provides the technology but does not operate the infrastructure
- Required by Chinese regulations — data stays within China
- Regions: China East, China North, China East 2, China North 2

### Azure confidential computing
- Hardware-level isolation using Intel SGX or AMD SEV-SNP enclaves
- Data is encrypted **even while being processed** — not just at rest or in transit
- Relevant for financial services, healthcare, and multi-party computation

## Azure for On-Premises: Arc & Stack

For organisations that cannot fully move to the cloud, Azure extends its control plane on-premises.

### Azure Arc
Projects Azure management capabilities onto resources running **anywhere** — on-premises, other clouds, or edge locations.
- Manage on-premises VMs, Kubernetes clusters, and databases through the Azure portal
- Apply Azure Policy, RBAC, and Microsoft Defender to non-Azure infrastructure
- No hardware shipped — it is a software agent installed on existing servers

### Azure Stack
Microsoft ships and installs physical hardware into your data centre, running a subset of Azure services locally.

| Product | Use case |
|---|---|
| **Azure Stack Hub** | Run Azure services in your own data centre — same APIs, same portal |
| **Azure Stack HCI** | Hyperconverged infrastructure for virtualised workloads on-premises |
| **Azure Stack Edge** | Edge compute appliance for AI inference and data processing at the network edge |

## Summary

| Component | What it is | Key detail |
|---|---|---|
| **Geography** | Country/compliance boundary | Contains ≥ 2 regions — data stays within |
| **Region** | Set of data centres in one location | 60+ worldwide |
| **Availability Zone** | Physically separate building within a region | Min. 3 per AZ-enabled region, < 2ms between zones |
| **Region Pair** | Two regions ≥ 300 miles apart in same geography | Staggered updates, failover priority, geo-replication target |
| **Availability Set** | FD + UD grouping within one data centre | Legacy HA — prefer AZs for new designs |
| **Edge Network** | CDN, Front Door, DNS edge nodes | Content and DNS acceleration globally |
| **Edge Zones** | Azure compute in metro cities / 5G networks | Low-latency scenarios |
| **Sovereign Regions** | Isolated clouds for government/regulated industries | Azure Government, Azure China |
| **Azure Arc** | Extend Azure management to anywhere | No hardware needed |
| **Azure Stack** | Azure hardware shipped to your data centre | Hub / HCI / Edge variants |

**The key principle:** Azure layers its infrastructure — Geographies for compliance, Regions for isolation, Availability Zones for fault tolerance, Region Pairs for disaster recovery — giving you multiple levels of protection to build on.
